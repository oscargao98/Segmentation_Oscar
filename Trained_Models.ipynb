```python
from lungmask import mask
import SimpleITK as sitk
from skimage.measure import label,regionprops
from skimage.segmentation import clear_border
import matplotlib.pyplot as plt
import os
```

```python
# Preparation

patient_id_str = './'+'ID00007637202177411956430' # example patient
if os.path.exists(patient_id_str + '.DS_Store'):
    os.remove(patient_id_str+'.DS_Store') # this MAC OS system file always messes things up
num_scans = len(os.listdir(patient_id_str))
print(num_scans)
```

# 1. UNET with lungmask Toolbox

### Careful: very slow with no GPU support

https://github.com/JoHof/lungmask

https://www.kaggle.com/smerllo/lung-mask-using-u-net-r231

```python
from lungmask import mask
import SimpleITK as sitk
from skimage.measure import label,regionprops
from skimage.segmentation import clear_border
import matplotlib.pyplot as plt
```

```python
for a_scan in range(num_scans):
    input_image = sitk.ReadImage(patient_id_str + '/' + str(a_scan+1) + '.dcm')
    segmentation = mask.apply(input_image)
    plt.imshow(segmentation[0])
    plt.show()
```
